```python
import pandas as pd
df_Carreras = pd.read_csv("DimCarrera.csv")
df_Carreras
```

```python
df_Estudiantes = pd.read_csv("DimEstudiantes.csv")
df_Estudiantes
```

```python
df_Periodo = pd.read_csv("DimPeriodo.csv")
df_Periodo
```

```python
df_Carreras.columns
```

```python
df_HechosPromedio = pd.read_json("HechosPromedio.json")
df_HechosPromedio
```

```python
df_inner1 = pd.merge(df_Carreras,df_Estudiantes, on='CodCarrera', how = 'inner')
df_inner1
```

```python
df_inner2 = pd.merge(df_Periodo,df_HechosPromedio, on='IdPeriodo', how = 'inner')
df_inner2
```

# Creacion de un unico Dataframe

```python
df_InfoGeneral = pd.merge(df_inner1,df_inner2, on='CuentaEstudiante', how = 'inner')
df_InfoGeneral
```

# Estudiante con mejor promedio en PrimerPAC

```python
df_PrimerPac = df_InfoGeneral[(df_InfoGeneral["NombrePeriodo"] == "PrimerPAC")]
df_PrimerPac
```

```python
df_PrimerPac[['CuentaEstudiante', 'PrimerNombre', 'PrimerApellido', 'NombreCarrera', 'Promedio']][df_PrimerPac.Promedio==df_PrimerPac.Promedio.max()]
```

# Estudiante con promedio mas bajo en el año 

```python
df_InfoGeneral[['CuentaEstudiante', 'PrimerNombre', 'PrimerApellido', 'NombreCarrera', 'Promedio']][df_InfoGeneral.Promedio==df_InfoGeneral.Promedio.min()]
```

# Estudiantes con excelencia academica en el año 

```python
df_Excelencia = df_InfoGeneral[['CuentaEstudiante', 'PrimerNombre', 'PrimerApellido', 'NombreCarrera', 'Promedio']][df_InfoGeneral["Promedio"]>=80]
```

```python
df_Excelencia.sort_values("Promedio", ascending=False)
```
